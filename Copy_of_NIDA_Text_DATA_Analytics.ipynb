```python
from marisa_trie import Trie
from pythainlp import word_tokenize
from pythainlp.corpus.common import thai_words
```

```python
#default word list from pythainlp
len(thai_words())
```

```python
add_words = ["‡∏õ‡∏±‡∏Å‡∏´‡∏°‡∏∏‡∏î","‡πÑ‡∏ü‡∏î‡∏±‡∏ö"]
custom_dict = list(thai_words())+add_words
len(custom_dict)
```

```python
trie_vocab = Trie(custom_dict)
```

```python
df['tokenized2'] = df.apply(lambda x : word_tokenize(
                                          x.comment, 
                                          keep_whitespace=False, 
                                          engine="newmm", 
                                          custom_dict=trie_vocab)
                    ,axis=1)
```

```python
df
```

```python
wlist = []
for d in df[:100].iterrows():
  tokenized = d[1].tokenized2
  wlist+=tokenized

word_count = Counter(wlist)
word_count.most_common(50)
```

```python
def get_news(keyword, site): 
    import time
    tcount = 1
    number_of_page = 10
    listresult = []
    # results = []
    for n in range(1,number_of_page+1): # from 1 to 5
        url = f'https://www.google.com/search?tbm=nws&start={n*10}&q={keyword} site:{site}' #default 10 title per result page
        res = requests.get(url,headers=headers) #get page's content
        # print(res)
        soup = BeautifulSoup(res.text, 'html.parser') #parse information from the page 
        # print(soup)
        hs = soup.findAll('div',{'role':'heading'}) #select targeted data
        for h in hs:
            print('-',tcount,h.text)
            tcount+=1
            listresult.append(h.text)
        time.sleep(1)
    return listresult
```

```python
res = requests.get(url,headers=headers) #get page's content
if res.status_code != 200:
    print("Error: Failed to get page")
```

```python
benz_res = get_news("Benz","pantip.com")
```

```python
BMW_res = get_news("BMW","pantip.com")
```

```python
toyota_res = get_news("toyota","pantip.com")
```

```python
import itertools
all_results = list(itertools.chain(benz_res, toyota_res,BMW_res))
print(len(all_results))
```

```python
dfx = pd.DataFrame(all_results,columns=['title'])
dfx
```
